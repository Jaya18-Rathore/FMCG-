```python
df['Month'] = pd.to_datetime(df["Month"])
df['month_name'] = df['Month'].dt.month_name(locale = 'English')
df['year_name'] = df ['Month'].dt.to_period('Y')
df['month_year'] = df['month_name']+' '+df['year_name'].astype(str)
df
```

```python
df['Qty'] = df['Qty'].str.replace(',','')
df['Sales'] = df['Sales'].str.replace(',','')
df
```

```python
df[["Sales", "Qty"]] = df[["Sales", "Qty"]].apply(pd.to_numeric)
df
```

1. Monthwise number of quantities sold

```python
month_qty = pd.DataFrame(df.groupby('month_name')['Qty'].sum().reset_index())
month_qty
```

2. Category wise Qty sold

```python
category_qty = pd.DataFrame(df.groupby('Category')['Qty'].sum().reset_index())
category_qty
```

3. Sub-channel wise total sales


```python
sub_channel_sales = pd.DataFrame(df.groupby('Sub-Channels')['Sales'].sum().reset_index())
sub_channel_sales
```

4. Highest sale in which month


```python
Higest_sale = pd.DataFrame(df.groupby('month_name')['Sales'].sum().reset_index())
Higest_sale 
```

5. Highest selling product (based on sales) (packaging quantity doesn't matter. Ex. consider "GET SLIM
JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)


```python
df['Product Name'] = df['Product Name'].str.replace('\d+','**')
df
```

```python
df['Product_1 Name'] = df['Product Name'].apply(lambda x:x.split("**")[0])
df
```

```python
Higest_selling_product = pd.DataFrame(df.groupby('Product_1 Name')['Sales'].sum().sort_values(['Sales']).reset_index())
Higest_selling_product
```

6. List of different products company is offering (packaging quantity doesn't matter. Ex. consider "GET
SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)

```python
diff_product = pd.DataFrame(df.groupby('Channel')['Product Name'].unique().rename())
diff_product
```
